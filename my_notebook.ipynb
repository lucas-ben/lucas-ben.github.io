```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/delinai/schulich_ds1/refs/heads/main/Datasets/kc_house_data.csv')
```

```python
df.info() # quick data quality check
# data 'object' want to switch to datetime
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.describe() # initial check
```

```python
df.iloc[:,9:].describe() 
```

```python
sns.histplot(df['price'], bins=50) # right skewed distribution
# most houses are less than $650,000
# after $1 million, the number of houses drops off
```

```python
sns.histplot(df['bedrooms'], bins=40)
```

```python
df[df['bedrooms']==33]
```

```python
df[(df['sqft_living'] > 1500) & (df['sqft_living'] < 1800) & (df['price'] > 600000) & (df['price'] < 700000)]['bedrooms'].value_counts()
# 33 bedrooms is an input mistake; learned from output (analytical support)
```

```python
df['bedrooms'] = df['bedrooms'].replace(33,3)
```

```python
df['bedrooms'].value_counts()
# want to check houses with 7,0,8,9,10,11 bedrooms seems suspicious
```

```python
df[df['bedrooms'] >= 7]['sqft_living'].median()
```

```python
df[df['bedrooms'] == 7]
```

```python
# top level trend; time series analysis
df.groupby('date')['price'].mean().plot()
```

```python
df.corr() # quick correlation analysis
```

```python
df['price_per_sqft'] = df['price'] / df['sqft_living'] # creating new variable
```

```python
df.groupby('date')['price_per_sqft'].median().plot()
```

```python
med_by_zipcode = df.groupby('zipcode')['price_per_sqft'].median()
```

```python
df['zipcode_med'] = df['zipcode'].map(med_by_zipcode)
```
