```python
import pandas as pd
df = pd.read_csv('baseball_practice_dataset.csv')
df.head()
df.columns
```

```python
import seaborn as sns
sns.regplot(x='hard_hit_percent',
            y='xwoba',
            data=df,
            ci=99)
```

```python
from statsmodels.formula.api import ols
model = ols('home_run ~ sweet_spot_percent + hard_hit_percent + xwoba', data=df).fit()
print(model.params)
print(model.summary())
```

```python
import matplotlib.pyplot as plt
sns.displot(data=df,
            x="home_run",
            col='year',
            col_wrap=2,
            bins=9)
plt.show()
```

```python
summary_stats = df.groupby('year')["home_run"].mean()
print(summary_stats)
```

```python
year_vs_hr = ols("home_run ~ year", data=df).fit()
print(year_vs_hr.params)
```

```python
# modelling without an intercept
year_vs_hr_no_intercept = ols("home_run ~ 0 + year", data=df).fit()
print(year_vs_hr_no_intercept.params)
```
